```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
import warnings
warnings.filterwarnings('ignore')
```

```python
data = {
    "Customer": ["Riya", "Aman", "Neha", "Faizan", "Imran", "Sneha"],
    "Age": [20, 22, 24, 26, 28, 23],
    "Spending": [100, 150, 210, 220, 250, 130]
}
```

```python
df = pd.DataFrame(data)
df
```

```python
X = df[["Age", "Spending"]]
X
```

```python
model = KMeans( n_clusters=2, n_init=10, random_state=42,)
model
```

```python
df["Group"] = model.fit_predict(X)
```

```python
df
```

```python
plt.figure(figsize=(6,5))
for group in df['Group'].unique():
    group_data = df[df['Group'] == group]
    plt.scatter(group_data["Age"], group_data["Spending"], label=f"Group {group}")

plt.xlabel("Age")
plt.ylabel("Spending")
plt.title("Customer Segements (K-Means)")
plt.legend()
plt.grid(True)
plt.show()
```
