```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
import joblib
```

```python
try:
    model = joblib.load('Student_score.pkl')
    scaler = joblib.load('scaler.pkl')
    print("model & scaler load sucssefull")
except FileNotFoundError:
    print("File does not exist. ")
    exit()
except Exception as e:
    print(f"Error: {e}")
    exit()
```

```python
try:
    study_hrs = float(input("Enter Study Hours: "))
    attendance = float(input("Enter Attendance: "))
    past_score = float(input("Enter Past Score: "))
    internet = input("Are Using Internet (Yes/No): ").strip().lower()
    sleep_hours = float(input("Enter Sleep Hours: "))
except Exception as e:
    print(f"Error {e}")

new_student = pd.DataFrame({
    'Study Hours': [study_hrs],
    'Attendance': [attendance],
    'PastScore': [past_score],
    "Internet": [internet],
    'SleepHours': [sleep_hours]
})

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

new_student['Internet'] = le.fit_transform(new_student["Internet"])
new_student_scaled = scaler.transform(new_student)

prediction = model.predict(new_student_scaled)

result = "Pass" if prediction[0] == 1 else "Fail"
print(f"Acconding to you submit data you have to {result} in this exam")
```
